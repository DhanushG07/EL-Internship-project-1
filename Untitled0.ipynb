## Data loading

### Subtask:
Load the employee attrition dataset.


**Reasoning**:

```python
import pandas as pd

try:
    df = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition (1).csv')
    display(df.head())
    print(df.shape)
except FileNotFoundError:
    print("Error: 'WA_Fn-UseC_-HR-Employee-Attrition (1).csv' not found.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

## Data exploration

### Subtask:
Explore the loaded dataset to understand its characteristics.


**Reasoning**:
I need to explore the dataset's characteristics, including shape, data types, missing values, and distributions of key variables.



```python
# Examine the shape of the DataFrame
print("Shape of the DataFrame:", df.shape)

# Investigate data types
print("\nData Types:")
print(df.info())

# Check for missing values
print("\nMissing Values per column:")
print(df.isnull().sum())

# Analyze distributions of key numerical variables
print("\nDescriptive statistics for key numerical variables:")
numerical_cols = ['Age', 'MonthlyIncome', 'YearsAtCompany']
print(df[numerical_cols].describe())

# Frequency of categorical variables
print("\nFrequency of key categorical variables:")
categorical_cols = ['Department', 'JobRole', 'EducationField', 'Attrition']
for col in categorical_cols:
    print(f"\nFrequency of {col}:")
    print(df[col].value_counts())
```

## Data analysis

### Subtask:
Analyze the relationship between employee attrition and other variables.


**Reasoning**:
Group the requested calculations and analyses into a single code block for efficiency.  Calculate descriptive statistics for numerical features grouped by attrition, analyze frequencies of categorical features for both attrition groups, and calculate the correlation matrix, including the conversion of 'Attrition' to a numerical representation.



```python
# Grouped calculations and analyses

# 1. Descriptive statistics for numerical features grouped by 'Attrition'
numerical_features = ['Age', 'MonthlyIncome', 'YearsAtCompany', 'JobSatisfaction', 'WorkLifeBalance', 'YearsInCurrentRole', 'YearsSinceLastPromotion']
grouped_stats = df.groupby('Attrition')[numerical_features].agg(['mean', 'median', 'std', 'min', 'max', lambda x: x.quantile(0.25), lambda x: x.quantile(0.75)])
display(grouped_stats)


# 2. Analyze frequency of categorical features for both attrition groups
categorical_features = ['Department', 'JobRole', 'MaritalStatus', 'EducationField', 'BusinessTravel', 'OverTime']
for col in categorical_features:
    print(f"\nFrequency of {col} by Attrition:")
    display(df.groupby('Attrition')[col].value_counts(normalize=True).unstack(fill_value=0))


# 3. Correlation matrix
# Convert 'Attrition' to numerical (0 for 'No', 1 for 'Yes')
df['Attrition_Num'] = df['Attrition'].map({'No': 0, 'Yes': 1})

# Select numerical columns for correlation
numerical_cols_corr = numerical_features + ['Attrition_Num']

# Calculate the correlation matrix
correlation_matrix = df[numerical_cols_corr].corr()
display(correlation_matrix)

# Identify strong correlations with attrition
attrition_correlations = correlation_matrix['Attrition_Num'].drop('Attrition_Num')
display(attrition_correlations.sort_values(ascending=False))
```

## Data visualization

### Subtask:
Visualize the data to gain insights into the distribution of features and relationships between them.


**Reasoning**:
Visualize the data using histograms, bar charts, box plots, and scatter plots to gain insights into the distribution of features and relationships between them.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograms for numerical features
plt.figure(figsize=(15, 10))
numerical_features = ['Age', 'MonthlyIncome', 'YearsAtCompany', 'JobSatisfaction', 'WorkLifeBalance']
for i, col in enumerate(numerical_features):
    plt.subplot(2, 3, i + 1)
    sns.histplot(data=df, x=col, hue='Attrition', kde=True)
    plt.title(f'Distribution of {col} by Attrition')
plt.tight_layout()
plt.show()


# Bar charts for categorical features
plt.figure(figsize=(15, 10))
categorical_features = ['Department', 'JobRole', 'MaritalStatus', 'EducationField', 'BusinessTravel', 'OverTime']
for i, col in enumerate(categorical_features):
    plt.subplot(3, 2, i+1)
    df.groupby('Attrition')[col].value_counts(normalize=True).unstack().plot(kind='bar', stacked=True)
    plt.title(f'{col} by Attrition')
plt.tight_layout()
plt.show()

# Box plots for numerical features
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_features):
    plt.subplot(2,3, i+1)
    sns.boxplot(data=df, x='Attrition', y=col)
    plt.title(f'Distribution of {col} by Attrition')
plt.tight_layout()
plt.show()

# Scatter plots for pairs of numerical variables
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
sns.regplot(data=df, x='MonthlyIncome', y='YearsAtCompany', hue='Attrition')
plt.subplot(2, 2, 2)
sns.regplot(data=df, x='Age', y='YearsAtCompany', hue='Attrition')
plt.subplot(2, 2, 3)
sns.regplot(data=df, x='JobSatisfaction', y='WorkLifeBalance', hue='Attrition')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograms for numerical features
plt.figure(figsize=(15, 10))
numerical_features = ['Age', 'MonthlyIncome', 'YearsAtCompany', 'JobSatisfaction', 'WorkLifeBalance']
for i, col in enumerate(numerical_features):
    plt.subplot(2, 3, i + 1)
    sns.histplot(data=df, x=col, hue='Attrition', kde=True)
    plt.title(f'Distribution of {col} by Attrition')
plt.tight_layout()
plt.show()


# Bar charts for categorical features
plt.figure(figsize=(15, 10))
categorical_features = ['Department', 'JobRole', 'MaritalStatus', 'EducationField', 'BusinessTravel', 'OverTime']
for i, col in enumerate(categorical_features):
    plt.subplot(3, 2, i+1)
    df.groupby('Attrition')[col].value_counts(normalize=True).unstack().plot(kind='bar', stacked=True)
    plt.title(f'{col} by Attrition')
plt.tight_layout()
plt.show()

# Box plots for numerical features
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_features):
    plt.subplot(2,3, i+1)
    sns.boxplot(data=df, x='Attrition', y=col)
    plt.title(f'Distribution of {col} by Attrition')
plt.tight_layout()
plt.show()

# Scatter plots for pairs of numerical variables using lmplot
sns.lmplot(data=df, x='MonthlyIncome', y='YearsAtCompany', hue='Attrition', height=5)
plt.title('Monthly Income vs Years at Company')
plt.show()

sns.lmplot(data=df, x='Age', y='YearsAtCompany', hue='Attrition', height=5)
plt.title('Age vs Years at Company')
plt.show()

sns.lmplot(data=df, x='JobSatisfaction', y='WorkLifeBalance', hue='Attrition', height=5)
plt.title('Job Satisfaction vs Work Life Balance')
plt.show()
```
